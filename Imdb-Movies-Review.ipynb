```python
# Define models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Support Vector Regression": SVR(),
    "K-Nearest Neighbors": KNeighborsRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42)
}

# Train and evaluate models
results = {}
for name, model in models.items():
    model.fit(x_train, y_train)  # Train model
    y_pred = model.predict(x_test)  # Make predictions
    rmse = mean_squared_error(y_test, y_pred, squared=False)  # Compute RMSE
    results[name] = rmse

# Sort models by RMSE (lower is better)
sorted_results = dict(sorted(results.items(), key=lambda item: item[1]))
sorted_results
```

```python
import numpy as np

# Train models and get predictions
model_predictions = {}
for name, model in models.items():
    model.fit(x_train, y_train)  # Train
    y_pred = model.predict(x_test)  # Predict
    model_predictions[name] = y_pred

# Create subplots for Actual vs Predicted Ratings
fig, axes = plt.subplots(1, 5, figsize=(18, 5))

for idx, (name, y_pred) in enumerate(model_predictions.items()):
    sns.scatterplot(x=y_test, y=y_pred, ax=axes[idx], alpha=0.5, color="blue")
    axes[idx].plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color="red", linestyle="dashed")  # Ideal line
    axes[idx].set_title(f"Actual vs Predicted ({name})")
    axes[idx].set_xlabel("Actual Rating")
    axes[idx].set_ylabel("Predicted Rating")

plt.tight_layout()
plt.show()

# Bar Plot for RMSE Comparison
plt.figure(figsize=(8, 5))
sns.barplot(x=list(sorted_results.keys()), y=list(sorted_results.values()), palette="coolwarm")
plt.title("Model RMSE Comparison")
plt.ylabel("Root Mean Squared Error (RMSE)")
plt.xlabel("Model")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [50, 100],  
    "max_depth": [None, 10],  
    "min_samples_split": [2, 5],  
    "min_samples_leaf": [1, 2]  
}

grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    cv=2,  # Reduce cross-validation folds
    scoring="neg_root_mean_squared_error",
    n_jobs=1  # Use only 1 CPU core
)

grid_search.fit(x_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best RMSE:", -grid_search.best_score_)
```
